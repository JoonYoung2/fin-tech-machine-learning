### relplot

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.relplot(x="total_bill", y="tip", data=tips) # x 증가 y 증가 = 양해 상관관계
```

```python
sns.relplot(x="total_bill", y="tip", hue="day",data=tips) # hue = 날짜별 그룹으로 알려줌
```

```python
sns.relplot(x="total_bill", y="tip", hue="day",data=tips, col="day") # col = 요일별로 분류해서 보여줌
```

```python
g = sns.scatterplot(x="total_bill", y="tip", hue="day",data=tips) # scatterplot속성에서는 col을 쓸 수 없음
g.set_title("scatter plot", fontsize=20, color="blue")
```

```python
plt.subplots( figsize=(10,3) )
sns.lineplot(x="total_bill", y="tip", hue="day",data=tips)
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 4))
g = sns.distplot(x=titanic['age'], color="darkblue", bins=10)
ax.set_title("DistPlot", fontsize=15)
#g, ax, plt 어떤거 써도 상관 없음
```

```python
fig, ax = plt.subplots(figsize=(10, 4))
g = sns.histplot(x=titanic['age'], color="skyblue", bins=10)
ax.set_title("HistPlot", fontsize=15)
```

```python
plt.subplots(figsize=(10, 3))
sns.boxplot( x = "day", y = "total_bill", hue = "smoker", data = tips )
```

```python
# 상관관계를 표현해준다
# 0.7이상이면 어느정도 관게가 있으며 0.3이하이면 관계가 없다고 표현
# 데이터 양이 많고 컬럼이 많을 때 사용
tips.corr( numeric_only=True )
```
